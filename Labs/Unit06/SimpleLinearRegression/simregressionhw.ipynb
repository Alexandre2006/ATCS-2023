The provided data from the advertising dataset describes the number of units of a product sold given the various TV, radio, and newspaper advertising budgets (in thousands).

## 1. Library Imports and Data Loading

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/advertising.csv")
```

## 2. Understanding the Data

```python
# Determine the features and label of the dataset
df.info()

# Graph x vs sales
sns.scatterplot(x="TV", y="sales", data=df)
plt.show()
sns.scatterplot(x="radio", y="sales", data=df)
plt.show()
sns.scatterplot(x="newspaper", y="sales", data=df)
plt.show()
```

## 3. Model 1

```python
# Create and train your model
X_train_tv = df[['TV']]
Y_train = df['sales']

model_tv = LinearRegression()
model_tv.fit(X_train_tv, Y_train)
```

```python
# Make the 2 predictions asked of you
model_tv.predict([[12], [345]])
```

```python
# Plot your data and line of best fit
# Don't forget to include a graph title with plt.title(...)
sns.scatterplot(data=df, x='TV', y='sales')
plt.plot(X_train_tv, model_tv.predict(X_train_tv), color='red')
plt.title('Sales vs TV advertising')
```

## 4. Model 2

```python
# Create and train your model
X_train_radio = df[['radio']]

model_radio = LinearRegression()
model_radio.fit(X_train_radio, Y_train)
```

```python
# Make the 2 predictions asked of you
model_radio.predict([[12], [345]])
```

```python
# Plot your data and line of best fit
# Don't forget to include a graph title with plt.title(...)
sns.scatterplot(data=df, x='radio', y='sales')
plt.plot(X_train_radio, model_radio.predict(X_train_radio), color='red')
plt.title('Sales vs Radio advertising')
```

## 5. Model 3

```python
# Create and train your model
X_train_newspaper = df[['newspaper']]

model_newspaper = LinearRegression()
model_newspaper.fit(X_train_newspaper, Y_train)
```

```python
# Make the 2 predictions asked of you
model_newspaper.predict([[12], [345]])
```

```python
# Plot your data and line of best fit
# Don't forget to include a graph title with plt.title(...)
sns.scatterplot(data=df, x='newspaper', y='sales')
plt.plot(X_train_newspaper, model_newspaper.predict(X_train_newspaper), color='red')
plt.title('Sales vs Newspaper advertising')
```
