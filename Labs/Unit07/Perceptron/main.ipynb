# Training a Perceptron on Linearly Seperable Data

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from perceptron import Perceptron
```

## 2. Load and Understand the Data

```python
# Load into a dataframe
df = pd.read_csv("data/bagel.csv")
```

```python
df.head()
```

```python
# View Data to see if it's linearly seperable
sns.scatterplot(data=df, x="radius", y="diameter", hue="bagel")
```

```python
# Split into features and label (X and y)
X, y = (df[["radius", "diameter"]].values, df["bagel"].values)

# Train-test split
X_train, X_test, y_train, y_test = (X[:400], X[400:], y[:400], y[400:])
```

## 3. Create the Model

```python
# Create the Model
model = Perceptron(2, epochs=300)

# Train the model
model.train(X_train, y_train)

# Get predictions
y_pred = model.predict(X_test)
```

## 4. Evaluate the Model

```python
# Print the weights
print(model.weights, model.bias)
```

```python
# Check the accuracy
model.score(y_test, y_pred)
```
